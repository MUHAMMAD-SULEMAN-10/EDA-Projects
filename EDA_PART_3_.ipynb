##Central limit thooram

###Central Limit Theorem

When you take many random samples from any population (doesn't matter if the population is normal, skewed, weird, or uneven), and calculate the average of each sample…

Those sample averages will form a normal (bell-shaped) distribution.

And…

The more samples you take, the more perfectly bell-shaped it becomes.

#confidence interval and margin of error

##confidence interval

A confidence interval (CI) is a range of values that is likely to contain the true value of something you’re trying to estimate.

A confidence interval tells you:

“We are fairly confident that the real value lies somewhere between these two numbers.”


##example

Suppose you measure the average height of a small group of people and get:

Sample mean: 170 cm

95% Confidence Interval: 168 cm to 172 cm

Meaning:

“We are 95% confident the true average height of the whole population is between 168 and 172 cm.”


## What does the “95% confidence” mean?

It does NOT mean there’s a 95% chance the true value is in your interval.

It means:

If you repeated the study 100 times,

about 95 out of those 100 intervals would contain the true value.

#point estimate


A point estimate is a single value that you use as the best guess for population parameter (like mean, proportion, etc.) based on your sample data.

##Explanation

You take a sample from a population.

You calculate something (like average, proportion, etc.) from the sample.

That value is your point estimate for the population




## Example

Suppose you survey 100 people about their height:

Sample mean height: 170 cm

Here:

170 cm is the point estimate of the true average height of the whole population.

 Key Points

It’s just one number (not a range).

Often paired with margin of error or confidence interval to show uncertainty.

#Margin of Error (MoE)

Margin of Error (MoE) simply tells you how much your sample estimate might differ from the true population value.

Margin of error is the amount of uncertainty around your estimate.

It shows how far your estimate could be from the real value.



#example

Suppose you survey people and find:

Sample mean salary: $50,000

Margin of error: ± $2,000

This means:

 The true average salary is likely between $48,000 and $52,000.

```python
from numpy.random import randn
population=randn(100)

```

```python
print(f" here is a mean of population: {np.mean(population)}, here is a std of populaiton: {np.std(population)}")
```

```python
null_mean= 0.022 # its a null hypothesis
```

```python
ALPHA = 0.05 # it will decide the domain expertise
```

```python
#CONFIDENCE INTERVAL = 95%
#Alpha value = 0.05(5%)

#import the statsmodel.stats.weightstats import ztest


# if p<0.05 we are going to regect the null hypothiss

# if p>0.05 we are going to accept the null hypothesis

#or

#P<=ALPHA
#p>ALPHA
```

```python
from statsmodels.stats.weightstats import ztest

z_score, p_value = ztest(population, value=null_mean, alternative='larger') #larger mean left and rigth side
```

```python
p_value # represents area
```

```python
#make a conclusion

if (p_value<ALPHA):
  print("reject the null hypothesis")
else:
  print("Accept the null hypothesis")
```
