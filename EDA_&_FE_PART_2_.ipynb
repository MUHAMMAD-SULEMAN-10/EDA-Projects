```python

# Categorical columns
category = ['Type', 'Content Rating']

plt.figure(figsize=(20, 15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.02)

for i in range(len(category)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=df[category[i]], color='skyblue')  # use 'color' instead of 'palette'
    plt.xlabel(category[i])
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# what are the top 5 most installed Apps in Each Popular Categories ??

#its a multivariant

# install--apps--category

df.head()
```

```python
dfa=df.groupby(['Category', 'App'])['Installs'].sum().reset_index()
```

```python
dfa
```

```python
dfa=dfa.sort_values('Installs', ascending=False)
```

```python
dfa
```

```python
#these are the most 5 toped installs apps

dfa[dfa.Category=='GAME'][:5]
```

```python
#check with respect to the family

dfa[dfa.Category=='FAMILY'][:5]
```

```python
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = dfa[dfa.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,x= 'Installs' ,y='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)

plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```
