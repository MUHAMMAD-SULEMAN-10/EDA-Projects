##Summary

All categories in the column you pass to OneHotEncoder are converted into numeric columns.

This happens for every categorical column individually if you encode them one by one.

Passing multiple columns at once will expand all categories in all columns into numeric columns.

The numbers you see (0/1) just indicate whether the row belongs to that category.

#fit( )
ohe.fit(df[['Airline']])


fit( ) learns all the unique categories from your data.

For Airline, it finds something like:



#output


['IndiGo', 'Air India', 'SpiceJet', 'Vistara', 'GoAir']

It does NOT yet convert anything. It just memorizes the categories and their order.

This is important when you want to apply the same encoding to new/unseen data.



#transform( )

encoded = ohe.transform(df[['Airline']])


transform( ) converts the actual data into numeric 0/1 columns using the

categories learned in fit( ).

##Example:

Original Airline column: ['IndiGo', 'Air India', 'SpiceJet']

Transform → One-hot numeric array:

IndiGo:       [1, 0, 0]
Air India:    [0, 1, 0]
SpiceJet:     [0, 0, 1]




##mportant:


transform( ) only works after fit( ) because it needs the mapping of category → column index

# fit_transform( )

encoded = ohe.fit_transform(df[['Airline']])


Combines both steps in one line:

fit( ) → learn the categories

transform( ) → convert your data into numeric columns

This is very convenient for training data.

##When to use fit( ) vs transform ( )

Training data: use fit_transform( ) → learn categories and encode at the same time.

Test or new data: use only transform( ) → apply the same mapping learned from training data.

Example:

# Training data
encoded_train = ohe.fit_transform(train_df[['Airline']])

# Test data
encoded_test = ohe.transform(test_df[['Airline']])


This ensures the same column order is used for train and test data.
